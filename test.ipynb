# BeautifulSoup情報を取得する

```python
url = "https://www.musashino-u.ac.jp/"
res = requests.get(url, headers=headers)
res.encoding = res.apparent_encoding
print(f"ステータスコード：{res.status_code}")
```

```python
#HTMLのソースコードを解析
soup = BeautifulSoup(res.text, 'html.parser')
print(f"soupの型：{type(soup)}")  # BeautifulSoupオブジェクトの型を表示
```

```python
# ソースコード全体から<a>タグを抽出
soup_a = soup.a
print(f"<a>タグ：{soup_a}")  # 最初に見つかった<a>タグを表示
print(f"<a>タグの型：{type(soup_a)}")  # <a>タグの型を表示

#ソースコード全体から<p>タグを抽出
soup_p = soup.p
print(f"<p>タグ：{soup_p}")  # 最初に見つかった<p>タグを表示
print(f"<p>タグの型：{type(soup_p)}")  # <p>タグの型を表示
```

```python
soup_div_div = soup.div.div
print(f"<div>タグの中の<div>タグ：{soup_div_div}")  # 最初に見つかった<div>タグを表示
print(f"<div>タグの中の<div>タグの型：{type(soup_div_div)}")  # <div>タグの型を表示
```

```python
# tagオブジェクト.stringでタグの中の文字列を取得
print(f"<a>タグのテキスト：{soup_a.string}")  # <a>タグのテキストを表示

# コンテンツがない場合はNoneが返る
print(f"<a>タグの中の<div>タグのテキスト：{soup_div_div.string}")  # <div>タグのテキストを表示
```

## 指定してHTMLタグで要素を検索

```python
#指定したタグが最初に使われている要素のうち、最初に見つかった要素を取得する。
h2 = soup.find('h2')
print(f"{h2}")  # 最初に見つかった<h2>タグを表示
print(f"<h2>変数の型：{type(h2)}")  # <h2>タグオブジェクトの型を表示
```

```python
h2_all = soup.find_all('h2')
#print(f"<h2>タグのリスト{h2_all}")  # <h2>タグのリストを表示
#print(f"<h2>型：{type(h2_all)}") #型を表示

for h2 in h2_all:
    if h2.span is not None:
        print(f"<h2>タグのテキスト：{h2.span.string}")  # <h2>タグのテキストを表示
```

```python
#a_list = soup.find_all('a')
#for a in a_list:
    #print(f"{a.string}")  # <a>タグ内のテキストを表示


    #検索結果の数に制限をかける
a_list_limited = soup.find_all('a', limit=5)
for a in a_list_limited:
    print(f"{a.string}")  # <a>タグ内のテキストを表示
```

```python
#複数のタグを指定して検索することも可能
#例えば、<h2>タグと<h3>タグの両方を検索する場合は以下のようにする。
h2_h3_list = soup.find_all(['h2', 'h3'] , limit=5)
for tag in h2_h3_list:
    print(f"{tag}") #<h2>タグと<h3>タグを表示
```

```python
tag_href_list = soup.find_all(attrs={'href': '/'}, limit=5)
for tag in tag_href_list:
    print(f"{tag}")  # href属性が'/'のタグを表示
```

```python
# id属性やclass属性など、任意の属性で検索することも可能
# 例えば、id属性が'top'の要素を検索する場合は以下
tag_id_list = soup.find_all(id='top')
for tag in tag_id_list:
    print(f"{tag}")  # id属性が'top'のタグを表示


tag_class_list = soup.find_all(class_='p-notices' , limit=5)
for tag in tag_class_list:
    print(f"{tag}")  # class属性が'p-notices'のタグを表示
```

## CSSセレクタを使った検索

```python
#id属性を指定（＃＋検索ID名）
id_tags = soup.select('#top')
for tag in id_tags:
    print(f"{tag}")# id属性が'top'のタグを表示


## class属性を使った検索
class_tags = soup.select('.p-notices')
for tag in class_tags:
    print(f"{tag}")# class属性が'p-notices'のタグを表示

## HTML階層構造を指定
#　例えば、<div>タグの中にある<a>タグを検索する場合は以下のようにする。
div_h2_tags = soup.select('div h2')
for tag in div_h2_tags:
    print(f"{tag}")  # <div>タグの中にある<h2>タグを表示
```

# 練習問題
下記のページから出演映画のタイトルを取得する
https://abehiroshi.la.coocan.jp/movie/eiga.htm

```python
#アクセスしたいwebサイトのURLを指定
url = 'https://abehiroshi.la.coocan.jp/movie/eiga.htm'

#webサーバーにHTTPリクエストを送信
#レスポンスに変数を格納しておく
res = requests.get(url)
```

```python
res.encoding = res.apparent_encoding
print(f"レスポンス:{res}")  # レスポンスオブジェクトを表示
print(f"レスポンスの型：{type(res)}")  # レスポンスオブジェクトの型を表示
print(f"ステータスコード：{res.status_code}")  # ステータスコードを表示
print(f"ステータスメッセージ：{res.reason}")  # ステータスメッセージを表示

print(f"リクエスト：{res.request}")  # リクエストオブジェクトを表示
print(f"リクエストの型：{type(res.request)}")  # リクエストオブジェクトの型を表示
print(f"リクエストヘッダー：{res.request.headers}")  # リクエストヘッダーを表示
print(f"リクエストメソッド：{res.request.method}")  # リクエストメソッドを表示
print(f"リクエストURL：{res.request.url}")  # リクエストURLを表示
```

```python
#レスポンス
print(f"レスポンスヘッダー：{res.headers}")  # レスポンスヘッダーを表示
print(f"レスポンスボディ：{res.text}")  # レスポンスボディを表示
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
```

```python
# 阿部寛のサイトにリクエストを送信（ヘッダー付き）
res = requests.get(url, headers=headers)
res.encoding = res.apparent_encoding

# HTMLを解析
soup = BeautifulSoup(res.text, 'html.parser')

# 映画のタイトルが含まれるテーブルの行を取得
movie_rows = soup.find_all('tr')

# 映画のタイトルを格納するリスト
movie_titles = []

# 各行から映画のタイトルを抽出
for row in movie_rows:
    # tdタグを含む行のみを処理
    title_cell = row.find('td')
    if title_cell:
        # タイトルのテキストを取得し、空白を削除
        title = title_cell.get_text().strip()
        if title:  # 空でない場合のみリストに追加
            movie_titles.append(title)

# 映画のタイトルを表示
print("阿部寛出演の映画タイトル一覧:")
for i, title in enumerate(movie_titles, 1):
    print(f"{i}. {title}")
```
